```python
import cv2
import numpy as np
import sys 
from PIL import Image
#img = Image.open(sys.argv[1])
#img= np.array(img)
#filename = sys.argv[1]
filename='tiles1.JPG'
name= filename[0:5]
img= cv2.imread(filename)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


edges = cv2.Canny(gray, 50, 150)

lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=50, maxLineGap=250)

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)


cv2.imwrite('robolin-'+name+'.jpg', img)
```
