```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
#filename= str(sys.argv[1])
#image = cv2.imread(filename)
# Read in the image
image= cv2.imread('location.png')
#image= image[40:96,15:70]  #building
#image= image[0:67,155:240]  #grass
image= image[140:250:,75:145] #road
plt.imshow(image)
# convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

pixel_vals = image.reshape((-1,3))
# Converting to float type
pixel_vals = np.float32(pixel_vals)

feature1=[]
feature2=[]
for i in range(len(pixel_vals)):
  r=pixel_vals[i][0]
  g=pixel_vals[i][1]
  b=pixel_vals[i][2]
  f1=abs(g-b)
  f2=abs(2*g-r-b)
  feature1.append(f1)
  feature2.append(f2)

feature1= np.asarray(feature1)
feature2= np.asarray(feature2)


b=sum(feature1) / len(feature1)
c=sum(feature2) / len(feature2)

if b<=6 and c>=1 and c<=8:
  label= 1
if b<=81 and c>=12 and c<=86:
  label= 2
if b>6 and b<=20 and c<=12:
  label= 3

print(label)
```
