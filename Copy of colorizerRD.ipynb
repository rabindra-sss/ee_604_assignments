```python
import cv2
import numpy as np
import matplotlib as plt
import sys




file1= str(sys.argv[1])
Y = cv2.imread(file1)

file2= str(sys.argv[2])
cb4 = cv2.imread(file2)

file3= str(sys.argv[3])
cr4 = cv2.imread(file3)

dim = (960,622)

# resize image
cb4 = cv2.cvtColor(cb4, cv2.COLOR_BGR2GRAY)
cb4 = cv2.resize(cb4, dim, interpolation = cv2.INTER_AREA)
cb4  = cv2.bilateralFilter(cb4, 7,80,80)

cr4 = cv2.cvtColor(cr4, cv2.COLOR_BGR2GRAY)
cr4 = cv2.resize(cr4, dim, interpolation = cv2.INTER_AREA)
cr4  = cv2.bilateralFilter(cr4, 7,80,80)

Y[:,:,1]= cb4
Y[:,:,2]= cr4



Y = cv2.cvtColor(Y, cv2.COLOR_YCrCb2RGB)

cv2.imwrite("flyingelephant.jpg", Y)
```
